```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

dados = {
    'Valor': [200, 220, 300, 290, 450, 457, 500, 530, 700, 800],
    'Idade': [18, 22, 23, 30, 35, 44, 49, 50, 67, 75]
}

dados =  pd.DataFrame(dados)

X = dados['Idade'].values
Y = dados['Valor'].values
X = X.reshape(-1,1)

regressor = LinearRegression()

regressor.fit(X,Y)

plt.scatter(X,Y,color = 'black')
plt.plot(X,regressor.predict(X),color ='red')
plt.show()

print(regressor.predict(np.array(57).reshape(-1,1)))
```

```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression


def valorReshape(n):
    return np.array(n).reshape(-1,1)

data = {"Area":[40,45,50,53,60,65,100,110,113,130],
        "Valor": [120,180,190,187,195,200,300,320,305,400]
        }

data = pd.DataFrame(data)

regressor = LinearRegression()

X = data["Area"].values.reshape(-1,1)
Y = data["Valor"].values

regressor.fit(X,Y)

plt.scatter(X,Y,color = 'blue')
plt.plot(X,regressor.predict(X),color="red")
plt.title("")

data.plot(kind='box')

print(f"Previsão para a area de 70 {regressor.predict(valorReshape(70))}")
print(f"Previsão para a area de 35 {regressor.predict(valorReshape(35))}")
print(f"Previsão para a area de 190 {regressor.predict(valorReshape(190))}")
```
