```python
#calculando PI com uma parte do circulo
import math
from  pickletools import read_int4
import numpy as np
import random
import matplotlib.pyplot as plt

horiz = np.array(range(100))/100.0
x_1 = np.ones(100)
plt.plot(horiz,x_1,'b')

vert = np.array(range(100))/100.0
y_1 = np.ones(100)
plt.plot(y_1,vert,'b')

n_in = 0
i = 1
N = 100000

while(i <= N):
    x = random.random()
    y = random.random()

    if(math.sqrt((x**2) + (y**2))) <= 1 :
        n_in += 1
        plt.plot(x,y,'go')
    
    else: #se o ponto estiver fora da circunferencia
        plt.plot(x,y,'ro')

    i += 1

pi = float((4*n_in)/N)
print(f"O valor de PI Ã© : {pi}")
plt.show()
```

```python
# calculando pi com a circunferencia toda
import numpy as np
import random 
import matplotlib.pyplot as plt
plt.style.use('bmh')

N = 500000
n_out =[]
n_in = []

for i in range(N):
    x = np.random.uniform(-1,1)
    y = np.random.uniform(-1,1)

    if np.sqrt(x**2 + y**2) <= 1 :
        n_in.append((x,y))
    else:
        n_out.append((x,y))

plt.figure(figsize=(6,6))

plt.scatter(
    [x[0] for x  in n_in],
    [y[1] for y in n_in],
    marker='.',
    alpha =0.5
)

plt.scatter(
    [x[0] for x in n_out],
    [y[1] for y in n_out],
    marker = '.',
    alpha = 0.5
)


pi = 4 * float(len(n_in)/N)

print(f'O valor de pi Ã©: {pi}')
plt.show()

```

```python
import numpy as np
from scipy.stats import binom 

jogadas = binom(n = 10, p = 0.5)

probabilidade_calculada = 1 - jogadas.cdf(3)

print('Valor obtido analiticamente por distribuiÃ§Ã£o Binomial')
print('Propabildiade de jgoa 10 vezes uma moeda e obtermos cara em mais que 3 jogadas', probabilidade_calculada)


def realiza_jogadas():
    jogadas = np.random.uniform(0,1,10)

    # print("Jogadas:", jogadas)

    # print('Soma das jogas com valores acima de 0.5 (50%):',(jogadas >0.5 ).sum() )
    return (jogadas >0.5).sum()


N = 100000
contagem = 0

for i in range(N):
    if(realiza_jogadas()) > 3:
        contagem +=1

probabilidade_calculada = float(contagem/N)

print("Valor obtido estocasticamente pelo metodo de Monte Carlo")
print('Propabildiade de jogar 10 a moeda e obtermos cara em mais de 3 jogas' , probabilidade_calculada)
```
