# 2)	Faça um algoritmo que trate os dados em um pré-processamento e realize três modelos de classificação. Ao final deve apresentar as métricas de erro com o melhor resultado possível. -> Prever se pessoa saiu do pais

```python
import pandas as pd
import matplotlib.pyplot as plt
from  sklearn.linear_model import LinearRegression
import sklearn.metrics
from sklearn.metrics import mean_absolute_error
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import confusion_matrix
from sklearn.metrics import accuracy_score, f1_score, precision_score, recall_score
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
```

```python

df = pd.read_csv("Churn_Modelling.csv")
df = df.dropna()
df = df.drop(columns=["CustomerId","Surname","RowNumber","Gender"])
df
```

```python

LABELS = df.columns  

encoder = LabelEncoder()
for col in LABELS:
    if df[col].dtype == 'O':
        df[col] = encoder.fit_transform(df[col])
    elif df[col].dtype == 'bool':
        df[col] = df[col].astype('int')

x = df.iloc[:,:-1].values
y = df.iloc[:,-1].values.reshape(-1,1)


x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.20)

```

# Logistic Regression

```python
clf = LogisticRegression()
clf.fit(x_train,y_train.ravel())

print(f"{'='*20} Decision Tree {'='*20}")
print(f"Score: {clf.score(x_test,y_test)}")
print("Matrix de confusao:")
print(confusion_matrix(y_test,clf.predict(x_test)))
print("--Metricas--")
print(f"Acuracia: {accuracy_score(y_test,clf.predict(x_test)) *100:.2f}%")
print(f"Precisao: {precision_score(y_test,clf.predict(x_test))*100:.2f}%")
print(f"Recall :{recall_score(y_test,clf.predict(x_test))*100:.2f}%")
print(f"F1 : {f1_score(y_test,clf.predict(x_test))*100:.2f}%")
```

# Decision Tree

```python
dtree = DecisionTreeClassifier()
dtree.fit(x_train,y_train.ravel())

print(f"{'='*20} Decision Tree {'='*20}")
print(f"Score: {dtree.score(x_test,y_test)}")
print("Matrix de confusao:")
print(confusion_matrix(y_test,dtree.predict(x_test)))
print("--Metricas--")
print(f"Acuracia: {accuracy_score(y_test,dtree.predict(x_test)) *100:.2f}%")
print(f"Precisao: {precision_score(y_test,dtree.predict(x_test))*100:.2f}%")
print(f"Recall :{recall_score(y_test,dtree.predict(x_test))*100:.2f}%")
print(f"F1 : {f1_score(y_test,dtree.predict(x_test))*100:.2f}%")
```

#  KNeighbors


```python
neigh = KNeighborsClassifier(n_neighbors=3)
neigh.fit(x_train,y_train.ravel())

print(f"{'='*20} Decision Tree {'='*20}")
print(f"Score: {neigh.score(x_test,y_test)}")
print("Matrix de confusao:")
print(confusion_matrix(y_test,neigh.predict(x_test)))
print("--Metricas--")
print(f"Acuracia: {accuracy_score(y_test,neigh.predict(x_test)) *100:.2f}%")
print(f"Precisao: {precision_score(y_test,neigh.predict(x_test))*100:.2f}%")
print(f"Recall :{recall_score(y_test,neigh.predict(x_test))*100:.2f}%")
print(f"F1 : {f1_score(y_test,neigh.predict(x_test))*100:.2f}%")
```
