```python
import pandas as pd
import numpy as np

grupo_1 = (1,8,10,38,39)
grupo_2= (8,10,39,45,49)

media_idade_grupo_1 = np.mean(grupo_1)
media_idade_grupo_2 = np.mean(grupo_2)

mediana_grupo_1 = np.median(grupo_1)
mediana_grupo_2 = np.median(grupo_2)

desvio_grupo_1 = np.std(grupo_1)
desvio_grupo_2 = np.std(grupo_2)

print(f'Media Grupo 1: {media_idade_grupo_1}')
print(f'Media Grupo 2: {media_idade_grupo_2}')

print(f'Mediana 1 :{mediana_grupo_1}')
print(f'Mediana 2 :{mediana_grupo_2}')

print(f'Desvio 1:{desvio_grupo_1}')
print(f'Desvio 2:{desvio_grupo_2}')
```

```python
# calcular a media das massas - j√° tenhooooo 

import pandas as pd
dados = {'Mistura 1': [22.02, 23.83, 26.67, 25.38, 25.49, 23.50, 25.90, 24.89],
         'Mistura 2': [21.49, 22.67, 24.62, 24.18, 22.78, 22.56, 24.46, 23.79],
         'Mistura 3': [20.33, 21.67, 24.67, 22.45, 22.29, 21.95, 20.49, 21.81]}

dados = pd.DataFrame(dados)

def CalculaDados(column):
    std = column.std()
    media = column.mean()
    mediana = column.median()
    return std,media,mediana


print('-'*15 + ' Mistura 1 ' + '-'*15)
dados_mist_1 = CalculaDados(dados['Mistura 1'])
print(f"Desvio: {dados_mist_1[0]} ")
print(f"Media: {dados_mist_1[1]}")
print(f"Mediana: {dados_mist_1[2]}")

print()

print('-'*15 + ' Mistura 2 ' + '-'*15)
dados_mist_2 = CalculaDados(dados['Mistura 2'])
print(f"Desvio: {dados_mist_2[0]} ")
print(f"Media: {dados_mist_2[1]}")
print(f"Mediana: {dados_mist_2[2]}")

print()

print('-'*15 + ' Mistura 3 ' + '-'*15)
dados_mist_3 = CalculaDados(dados['Mistura 3'])
print(f"Desvio: {dados_mist_3[0]} ")
print(f"Media: {dados_mist_3[1]}")
print(f"Mediana: {dados_mist_3[2]}")

boxplot = dados.boxplot()
```
