Em uma determinada escola de ensino médio, alguns alunos foram selecionados aleatoriamente para saber qual idade eles tem. Abaixo temos a amostra das idades desse alunos. Calcule a média e o desvio padrão, e elabore os gráficos de boxplot, histograma e a curva da distribuição normal.


```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import random
import datetime
from scipy.stats import norm
```

```python
dados = {"Idade": [random.randint(13, 21) for i in range(25)]}

# dados={"Idade": [14, 17 ,18, 15, 15, 16, 17, 15, 16, 16, 15, 17, 15, 16, 
# 16, 18, 18, 19, 17, 16, 17, 15, 16, 17, 17, 19, 20, 18, 
# 17, 16, 15, 16, 16, 17, 18, 18, 17, 17, 15, 16, 16, 15]}
df = pd.DataFrame(dados)
```

```python
media = np.mean(df["Idade"])
desvio = np.std(df["Idade"])
dominio = np.linspace(np.min(df["Idade"]), np.max(dados["Idade"]))


print(f"media: {media}")
print(f"Desvio: {desvio}")
```

```python
fig,ax = plt.subplots(nrows =2, ncols=2)
fig.suptitle("All Graphs")

ax[0][0].boxplot(df["Idade"])
ax[0][1].hist(df["Idade"])
ax[1][0].plot(dominio, norm.pdf(dominio,media,desvio))
```
