```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
import random as rd


dados = {'X': [5,15,25,35,45,55],
         'Y': [5,20,14,32,22,38]}

df = pd.DataFrame(dados)
X = df['X'].values
Y = df["Y"].values

X = X.reshape(-1,1)

ajuste1 = LinearRegression(fit_intercept=False)
ajuste2= LinearRegression(fit_intercept=True)

ajuste1.fit(X,Y)
ajuste2.fit(X,Y)

R_sq1 = ajuste1.score(X,Y)
R_sq2 = ajuste2.score(X,Y)

b0_ajuste1 = ajuste1.intercept_
b1_ajuste1 = ajuste1.coef_

b0_ajuste2 = ajuste2.intercept_
b1_ajuste2 = ajuste2.coef_

print("\n" + "="*15)
print("\n Paramentros obtidos para o ajuste 1")
print("\n" + "="*15)
print(f"\n Coeficiente Linear para ajuste 1: {b0_ajuste1}")
print(f"\n Coeficiente angular para ajuste 1: {b1_ajuste1}")
print(f"\n Valor Para o raio: {R_sq1}")
print("\n" + "="*15)

print("\n" + "="*15)
print("\n Paramentros obtidos para o ajuste 2")
print("\n" + "="*15)
print(f"\n Coeficiente Linear para ajuste 2: {b0_ajuste2}")
print(f"\n Coeficiente angular para ajuste 2: {b1_ajuste2}")
print(f"\n Valor Para o raio: {R_sq2}")
print("\n" + "="*15)


fig_size = plt.figure(figsize = (10,6))

plot_ajuste1 = fig_size.add_subplot(121)
plot_ajuste1 = plt.scatter(X,Y,color = 'blue')
plot_ajuste1 = plt.plot(X,ajuste1.predict(X), color = "red")
plot_ajuste1 = plt.xlabel("X")
plot_ajuste1 = plt.ylabel("Y")

plot_ajuste2 = fig_size.add_subplot(122)
plot_ajuste2 = plt.scatter(X,Y,color = 'blue')
plot_ajuste2 = plt.plot(X,ajuste2.predict(X), color = "red")
plot_ajuste2 = plt.xlabel("X")
plot_ajuste2 = plt.ylabel("Y")

plt.show()
```

```python

import numpy as np
from sklearn.linear_model import LinearRegression

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import statsmodels.formula.api as smf

x= [[2, 1], [5, 3], [15, 2], [25, 5], [35, 11], [45, 15], [55, 34], [60, 35],
    [64, 40], [70, 41], [78, 32], [83, 45], [90, 51], [95, 55], [100, 64], [104, 65]]

y= [4, 5, 20, 14, 32, 42, 38, 43, 36, 25, 50, 44, 52, 62, 68, 73]

ajuste1 = LinearRegression(fit_intercept=False)
ajuste2 = LinearRegression(fit_intercept=True)

ajuste1.fit(x,y)
ajuste2.fit(x,y)

R_sq1 = ajuste1.score(x,y)
R_sq2 = ajuste2.score(x,y)

b0_ajuste1 = ajuste1.intercept_
b1_ajuste1 = ajuste1.coef_[0]
b2_ajuste1 = ajuste1.coef_[1]

b0_ajuste2 = ajuste2.intercept_
b1_ajuste2 = ajuste2.coef_[0]
b2_ajuste2 = ajuste2.coef_[1]

print(f"\n {'='*60}")
print('** PARÂMETROS OBTIDOS PARA O AJUSTE 1 **')
print(f"{'='*60}")
print("Coeficiente Linear para o ajuste1 b0 : ", b0_ajuste1 )
print('Coeficiente angular b1 para o ajuste 1: ',b1_ajuste1)
print('Coeficiente angular b2 para o ajuste 1: ',b2_ajuste1)
print('Valor para R^2 para o ajuste 1: ', R_sq1)
print(f"{'='*60}")


print(f"\n {'='*60}")
print('** PARÂMETROS OBTIDOS PARA O AJUSTE 2 **')
print(f"{'='*60}")
print("Coeficiente Linear para o ajuste2 b0 : ", b0_ajuste2 )
print('Coeficiente angular b1 para o ajuste 2: ',b1_ajuste2)
print('Coeficiente angular b2 para o ajuste 2: ',b2_ajuste2)
print('Valor para R^2 para o ajuste 1: ', R_sq2)
print(f"{'='*60}")

df = pd.DataFrame(x, columns=['x1','x2'])
df['y'] = pd.Series(y)

ajuste_modelo = smf.ols(formula='y ~ x1 + x2', data=df)
equacao_ajustada = ajuste_modelo.fit()
equacao_ajustada.params

x_surf, y_surf = np.meshgrid(np.linspace(df.x1.min(),df.x1.max(),100), np.linspace(df.x2.min(),df.x2.max(),100))
onlyX = pd.DataFrame({'x1':x_surf.ravel(), 'x2': y_surf.ravel()})
fittedY= equacao_ajustada.predict(exog=onlyX)

fittedY = np.array(fittedY)

fig = plt.figure(dpi = 100)
ax = fig.add_subplot(111, projection = '3d')
ax.scatter(df['x1'], df['x2'], df['y'], c = 'red', marker = 'o', alpha=0.5)
ax.plot_surface(x_surf, y_surf, fittedY.reshape(x_surf.shape), color ='blue', alpha = 0.3)
plt.show()
```
